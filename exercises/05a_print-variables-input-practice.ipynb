## Exercise 2: Age in Dog Years

Create a program that calculates a person's age in "dog years":

1. Ask the user for their name
2. Ask the user for their age in human years
3. Convert the input to an integer
4. Calculate their age in dog years (the formula is: human age * 7)
5. Print a fun message showing both their human age and dog age

Example output: "Hi Maria! You are 25 in human years, which is 175 in dog years. Woof!"

````
# write your code below this line

# Ask for user's name and age
name = input("Enter your name: ")
human_age = input("Enter your age in human years: ")

# Convert age to integer
human_age = int(human_age)

# Calculate age in dog years
dog_age = human_age * 7

# Print the fun message
print(f"Hi {name}! You are {human_age} in human years, which is {dog_age} in dog years. Woof!")
````

## Exercise 4: Treasure Hunter

Create a program that calculates a treasure hunter's score:

1. Create variables for three types of treasures with their point values (use integers)
   - Gold coins: 10 points
   - Silver rings: 5 points
   - Bronze statues: 15 points
2. Ask the user how many of each treasure they found
3. Calculate the total score
4. Print out a detailed report showing quantities, points per item, and total score

Example output:
```
=== TREASURE HUNTER REPORT ===
Gold coins: 3 x 10 = 30 points
Silver rings: 2 x 5 = 10 points
Bronze statues: 1 x 15 = 15 points
----------------------------
Total score: 55 points
=============================
```

````
# write your code below this line

# Define treasures and their point values
gold_coin_points = 10
silver_ring_points = 5
bronze_statue_points = 15

# Get quantities from user
gold_coins = int(input("How many gold coins did you find? "))
silver_rings = int(input("How many silver rings did you find? "))
bronze_statues = int(input("How many bronze statues did you find? "))

# Calculate scores
gold_score = gold_coins * gold_coin_points
silver_score = silver_rings * silver_ring_points
bronze_score = bronze_statues * bronze_statue_points
total_score = gold_score + silver_score + bronze_score

# Print treasure report
print("=== TREASURE HUNTER REPORT ===")
print(f"Gold coins: {gold_coins} x {gold_coin_points} = {gold_score} points")
print(f"Silver rings: {silver_rings} x {silver_ring_points} = {silver_score} points")
print(f"Bronze statues: {bronze_statues} x {bronze_statue_points} = {bronze_score} points")
print("----------------------------")
print(f"Total score: {total_score} points")
print("=============================")
````

## Exercise 5: Secret Code Generator

Create a program that converts a message into a "secret code":

1. Ask the user for their name
2. Ask the user for a secret message
3. Create a "coded" version of their message by:
   - Reversing the message
   - Adding a special symbol (like "*") between each character
4. Print both the original message and the coded message

Example:
If the user enters "hello", your program should output "o*l*l*e*h"

````
# write your code below this line

# Get user's name and message
name = input("Enter your name: ")
message = input("Enter a secret message: ")

# Create the coded message
reversed_message = message[::-1]
coded_message = "*".join(reversed_message)

# Print both messages
print(f"Hello {name}!")
print(f"Original message: {message}")
print(f"Coded message: {coded_message}")
````

## Bonus Exercise: Tip Calculator

Create a tip calculator for restaurant bills:

1. Ask the user for the total bill amount
2. Ask the user for the desired tip percentage (15%, 18%, or 20%)
3. Convert inputs to floats
4. Calculate the tip amount and the total bill including tip
5. Print the tip amount and total bill in a nice format with 2 decimal places

Example output: 
```
Bill: $50.00
Tip (18%): $9.00
Total: $59.00
```

````
# write your code below this line

# Get bill amount and tip percentage
bill_amount = float(input("Enter the bill amount: $"))
tip_percentage = float(input("Enter the tip percentage (15, 18, or 20): "))

# Calculate tip and total
tip_amount = bill_amount * (tip_percentage / 100)
total_amount = bill_amount + tip_amount

# Print formatted bill 
print(f"Bill: ${bill_amount:.2f}")
print(f"Tip ({int(tip_percentage)}%): ${tip_amount:.2f}")
print(f"Total: ${total_amount:.2f}") 
````
