## Übung 2: Alter in Hundejahren

Erstelle ein Programm, das das Alter einer Person in "Hundejahren" berechnet:

1. Frage den Benutzer nach seinem Namen
2. Frage den Benutzer nach seinem Alter in Menschenjahren
3. Wandle die Eingabe in eine Ganzzahl um
4. Berechne sein Alter in Hundejahren (die Formel ist: Menschenalter / 7)
5. Gib eine lustige Nachricht aus, die sowohl das menschliche Alter als auch das Hundealter zeigt

Beispielausgabe: "Hallo Maria! Du bist 70 in Menschenjahren, was 10 in Hundejahren entspricht. Wuff!"

````python
# schreibe deinen Code unter dieser Zeile

# Frage nach dem Namen und Alter des Benutzers
name = input("Gib deinen Namen ein: ")
menschenalter = input("Gib dein Alter in Menschenjahren ein: ")

# Wandle das Alter in eine Ganzzahl um
menschenalter = int(menschenalter)

# Berechne das Alter in Hundejahren
hundealter = menschenalter / 7

# Gib die lustige Nachricht aus
print(f"Hallo {name}! Du bist {menschenalter} in Menschenjahren, was {hundealter} in Hundejahren entspricht. Wuff!")
````

## Übung 4: Schatzjäger

Erstelle ein Programm, das die Punktzahl eines Schatzjägers berechnet:

1. Erstelle Variablen für drei Arten von Schätzen mit ihren Punktwerten (nutze Ganzzahlen)
   - Goldmünzen: 10 Punkte
   - Silberringe: 5 Punkte
   - Bronzestatuen: 15 Punkte
2. Frage den Benutzer, wie viele von jedem Schatz er gefunden hat
3. Berechne die Gesamtpunktzahl
4. Gib einen detaillierten Bericht aus, der Mengen, Punkte pro Gegenstand und Gesamtpunktzahl zeigt

Beispielausgabe:
```
=== SCHATZJÄGER BERICHT ===
Goldmünzen: 3 x 10 = 30 Punkte
Silberringe: 2 x 5 = 10 Punkte
Bronzestatuen: 1 x 15 = 15 Punkte
----------------------------
Gesamtpunktzahl: 55 Punkte
===========================
```

````python
# schreibe deinen Code unter dieser Zeile

# Definiere Schätze und ihre Punktwerte
goldmuenze_punkte = 10
silberring_punkte = 5
bronzestatue_punkte = 15

# Erfasse Mengen vom Benutzer
goldmuenzen = int(input("Wie viele Goldmünzen hast du gefunden? "))
silberringe = int(input("Wie viele Silberringe hast du gefunden? "))
bronzestatuen = int(input("Wie viele Bronzestatuen hast du gefunden? "))

# Berechne Punktzahlen
gold_punktzahl = goldmuenzen * goldmuenze_punkte
silber_punktzahl = silberringe * silberring_punkte
bronze_punktzahl = bronzestatuen * bronzestatue_punkte
gesamtpunktzahl = gold_punktzahl + silber_punktzahl + bronze_punktzahl

# Gib Schatzbericht aus
print("=== SCHATZJÄGER BERICHT ===")
print(f"Goldmünzen: {goldmuenzen} x {goldmuenze_punkte} = {gold_punktzahl} Punkte")
print(f"Silberringe: {silberringe} x {silberring_punkte} = {silber_punktzahl} Punkte")
print(f"Bronzestatuen: {bronzestatuen} x {bronzestatue_punkte} = {bronze_punktzahl} Punkte")
print("----------------------------")
print(f"Gesamtpunktzahl: {gesamtpunktzahl} Punkte")
print("===========================")
````

## Übung 5: Geheimcode-Generator

Erstelle ein Programm, das eine Nachricht in einen "Geheimcode" umwandelt:

1. Frage den Benutzer nach seinem Namen
2. Frage den Benutzer nach einer geheimen Nachricht
3. Erstelle eine "kodierte" Version seiner Nachricht, indem du:
   - Ein spezielles Symbol (wie "*") zwischen jeden Buchstaben einfügst
   (Dafür kannst du folgenden Code verwenden: `"*".join(nachricht)`)
4. Gib sowohl die ursprüngliche Nachricht als auch die kodierte Nachricht aus

Beispiel:
Wenn der Benutzer "hallo" eingibt, sollte dein Programm `"h*a*l*l*o"` ausgeben

````python
# schreibe deinen Code unter dieser Zeile

# Erfasse Name und Nachricht des Benutzers
name = input("Gib deinen Namen ein: ")
nachricht = input("Gib eine geheime Nachricht ein: ")

# Erstelle die kodierte Nachricht
kodierte_nachricht = "*".join(nachricht)

# Gib beide Nachrichten aus
print(f"Hallo {name}!")
print(f"Ursprüngliche Nachricht: {nachricht}")
print(f"Kodierte Nachricht: {kodierte_nachricht}")
````

## Bonus-Übung: Trinkgeldrechner

Erstelle einen Trinkgeldrechner für Restaurantrechnungen:

1. Frage den Benutzer nach dem Gesamtrechnungsbetrag
2. Frage den Benutzer nach dem gewünschten Trinkgeldprozentsatz (15%, 18% oder 20%)
3. Wandle die Eingaben in Gleitkommazahlen um
4. Berechne den Trinkgeldbetrag und den Gesamtrechnungsbetrag inklusive Trinkgeld
5. Gib den Trinkgeldbetrag und die Gesamtrechnung in einem schönen Format mit 2 Dezimalstellen aus

Beispielausgabe:
```
Rechnung: 50,00 €
Trinkgeld (18%): 9,00 €
Gesamt: 59,00 €
```

````python
# schreibe deinen Code unter dieser Zeile

# Erfasse Rechnungsbetrag und Trinkgeldprozentsatz
rechnungsbetrag = float(input("Gib den Rechnungsbetrag ein: € "))
trinkgeldprozentsatz = float(input("Gib den Trinkgeldprozentsatz ein (15, 18 oder 20): "))

# Berechne Trinkgeld und Gesamtbetrag
trinkgeldbetrag = rechnungsbetrag * (trinkgeldprozentsatz / 100)
gesamtbetrag = rechnungsbetrag + trinkgeldbetrag

# Gib formatierte Rechnung aus
print(f"Rechnung: {rechnungsbetrag:.2f} €")
print(f"Trinkgeld ({int(trinkgeldprozentsatz)}%): {trinkgeldbetrag:.2f} €")
print(f"Gesamt: {gesamtbetrag:.2f} €") 
````
