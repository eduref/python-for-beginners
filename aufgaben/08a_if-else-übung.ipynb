## √úbung 1: Temperatur-Klassifikator

Erstelle ein Programm, das:
1. Den Benutzer nach einem Temperaturwert in Celsius fragt
2. Die Temperatur wie folgt klassifiziert:
   - "Gefrierend" wenn unter 0¬∞C
   - "Kalt" wenn zwischen 0¬∞C und 15¬∞C
   - "M√§√üig" wenn zwischen 15¬∞C und 25¬∞C
   - "Hei√ü" wenn zwischen 25¬∞C und 35¬∞C
   - "Sehr hei√ü" wenn √ºber 35¬∞C

````python
# schreibe deinen Code unter dieser Zeile

# Temperatur vom Benutzer erhalten
temperatur = float(input("Gib die Temperatur in Celsius ein: "))

# Temperatur klassifizieren
if temperatur < 0:
    print("Gefrierend")
elif temperatur < 15:
    print("Kalt")
elif temperatur < 25:
    print("M√§√üig")
elif temperatur < 35:
    print("Hei√ü")
else:
    print("Sehr hei√ü")
````

## √úbung 2: Passwort-St√§rke-Pr√ºfer

Erstelle ein Programm, das die St√§rke eines Passworts √ºberpr√ºft:

1. Fordere den Benutzer auf, ein Passwort einzugeben
2. Bewerte das Passwort und gib seine St√§rke basierend auf diesen Kriterien aus:
   - "Schwach" wenn es weniger als 8 Zeichen hat
   - "Mittel" wenn es 8 oder mehr Zeichen hat, aber nur Buchstaben (`isalpha`) oder nur Zahlen (`isnum`) enth√§lt
   - "Stark" wenn es 8 oder mehr Zeichen hat und sowohl Buchstaben als auch Zahlen enth√§lt (`isalnum` sagt, ob es **nur** Zahlen und Buchstaben hat)
   - "Sehr stark" wenn es 8 oder mehr Zeichen hat, Buchstaben, Zahlen und mindestens ein Sonderzeichen (!, @, #, $, %, usw.) enth√§lt

Hinweis: Du kannst die String-Methoden `passwort.isalpha()`, `.isdigit()` und `.isalnum()` verwenden, um den Inhalt des Passworts zu √ºberpr√ºfen.

Beispiel:
```
if password.isalpha():
   print("Das Passwort beinhaltet nur Buchstaben")
```

````python
# schreibe deinen Code unter dieser Zeile

# Passwort vom Benutzer erhalten
passwort = input("Gib dein Passwort ein: ")

# Passwort-St√§rke √ºberpr√ºfen
if len(passwort) < 8:
    st√§rke = "Schwach"
elif passwort.isalpha() or passwort.isdigit():
    st√§rke = "Mittel"
elif passwort.isalnum():
    st√§rke = "Stark"
else:
    st√§rke = "Sehr stark"

print(f"Passwort-St√§rke: {st√§rke}")
````

## √úbung 3: Kinokarten-Preisberechnung

Erstelle ein Programm, das Kinokarten-Preise basierend auf Alter und Wochentag berechnet:

1. Frage nach dem Alter der Person
2. Frage nach dem Wochentag (Montag, Dienstag, usw.)
3. Berechne und zeige den Ticketpreis basierend auf diesen Regeln an:
   - Grundpreis ist 12‚Ç¨
   - Kinder (unter 12) erhalten 50% Rabatt
   - Senioren (65 und √§lter) erhalten 30% Rabatt
   - Am Mittwoch bekommen alle 25% Rabatt (dieser wird nach altersbedingten Rabatten angewendet)
   - An Wochenenden (Samstag und Sonntag) kommt ein Aufschlag von 2‚Ç¨ hinzu (nach allen Rabatten)

Achte darauf, den Originalpreis und den Endpreis mit einer Aufschl√ºsselung der angewendeten Rabatte anzuzeigen.

````python
# schreibe deinen Code unter dieser Zeile

# Alter und Wochentag abfragen
alter = int(input("Gib dein Alter ein: "))
tag = input("Gib den Wochentag ein: ").lower()

# Ticketpreis berechnen
grundpreis = 12
endpreis = grundpreis
rabatt = 0
aufschlag = 0

print(f"Basis-Ticketpreis: {grundpreis}‚Ç¨")

# Altersbasierte Rabatte anwenden
if alter < 12:
    rabatt = grundpreis * 0.5
    endpreis -= rabatt
    print(f"Kinderrabatt (50%): -{rabatt}‚Ç¨")
elif alter >= 65:
    rabatt = grundpreis * 0.3
    endpreis -= rabatt
    print(f"Seniorenrabatt (30%): -{rabatt}‚Ç¨")

# Mittwochs-Rabatt anwenden
if tag == "mittwoch":
    mittwochs_rabatt = endpreis * 0.25
    endpreis -= mittwochs_rabatt
    print(f"Mittwochs-Rabatt (25%): -{mittwochs_rabatt:.2f}‚Ç¨")

# Wochenend-Aufschlag anwenden
if tag == "samstag" or tag == "sonntag":
    aufschlag = 2
    endpreis += aufschlag
    print(f"Wochenend-Aufschlag: +{aufschlag}‚Ç¨")

print(f"Endg√ºltiger Ticketpreis: {endpreis:.2f}‚Ç¨")
````

## √úbung 4: Schere, Stein, Papier Spiel

Erstelle ein einfaches Schere, Stein, Papier Spiel:

1. Bitte den Benutzer, Schere, Stein oder Papier zu w√§hlen
2. Lass den Computer zuf√§llig Schere, Stein oder Papier w√§hlen
3. Bestimme den Gewinner basierend auf den Regeln:
   - Stein schl√§gt Schere
   - Schere schl√§gt Papier
   - Papier schl√§gt Stein
   - Gleiche Wahlen f√ºhren zu einem Unentschieden
4. Zeige beide Wahlen und das Ergebnis an (Gewinn, Verlust oder Unentschieden)

Hinweis: Um eine zuf√§llige Wahl f√ºr den Computer zu generieren, kannst du folgendes verwenden:
```python

```

````python
import random

# Computer-Wahl generieren
computer_wahl = random.choice(["stein", "papier", "schere"])

# schreibe deinen Code unter dieser Zeile

# Benutzerwahl erfassen

benutzer_wahl = input("Gib deine Wahl ein (stein, papier, schere): ").lower()

# Gewinner bestimmen
print(f"Deine Wahl: {benutzer_wahl}")
print(f"Computer-Wahl: {computer_wahl}")

if benutzer_wahl == computer_wahl:
    print("Unentschieden!")
elif (benutzer_wahl == "stein" and computer_wahl == "schere") or \
     (benutzer_wahl == "schere" and computer_wahl == "papier") or \
     (benutzer_wahl == "papier" and computer_wahl == "stein"):
    print("Du gewinnst!")
else:
    print("Computer gewinnt!")
````

## √úbung 5: Zahlenraten-Spiel

Erstelle ein Zahlenraten-Spiel:

1. Generiere eine zuf√§llige Zahl zwischen 1 und 100
2. Fordere den Benutzer auf, die Zahl zu erraten
3. Teile ihm mit, ob sein Tipp zu hoch, zu niedrig oder korrekt ist
4. Lass ihn weiter raten, bis er die richtige Zahl findet
5. Behalte die Anzahl der Versuche im Auge und zeige sie an, wenn er gewinnt

Hinweis: Um eine Zufallszahl zu generieren, kannst du folgendes verwenden:
```python
import random
geheimzahl = random.randint(1, 100)
```

````python
# Geheimzahl generieren
import random
geheimzahl = random.randint(1, 100)

# schreibe deinen Code unter dieser Zeile

# Spielschleife zum Raten
versuche = 0
tipp = None

print("Ich denke an eine Zahl zwischen 1 und 100.")

while tipp != geheimzahl:
    tipp = int(input("Gib deinen Tipp ein: "))
    versuche += 1
    
    if tipp < geheimzahl:
        print("Zu niedrig! Versuche es noch einmal.")
    elif tipp > geheimzahl:
        print("Zu hoch! Versuche es noch einmal.")
    else:
        print(f"Herzlichen Gl√ºckwunsch! Du hast die Zahl in {versuche} Versuchen erraten!")
````
