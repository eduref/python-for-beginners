## Ãœbung 4: Warenkorb

Erstellen Sie ein Programm, das einen Warenkorb simuliert:

1. Erstellen Sie ein WÃ¶rterbuch mit dem Namen `products`, das mindestens 5 Artikel mit ihren Preisen enthÃ¤lt
   - Beispiel: `{"Apfel": 0.75, "Banane": 0.99, "Brot": 3.49, ...}`
2. Erstellen Sie eine leere Liste mit dem Namen `cart`
3. Bitten Sie den Benutzer, drei Artikel einzugeben, die er kaufen mÃ¶chte
4. FÃ¼r jeden Artikel:
   - ÃœberprÃ¼fen Sie, ob er im WÃ¶rterbuch `products` existiert
   - Wenn ja, fÃ¼gen Sie der Liste `cart` ein WÃ¶rterbuch mit dem Artikelnamen und dem Preis hinzu
   - Wenn nicht, geben Sie eine Meldung aus, dass der Artikel nicht verfÃ¼gbar ist
5. Nachdem alle Artikel hinzugefÃ¼gt wurden, geben Sie den Inhalt des Warenkorbs und die Gesamtkosten aus

Beispielausgabe:
```
Ihr Warenkorb:
- Apfel: 0,75 â‚¬
- Brot: 3,49 â‚¬
- Milch: 2,50 â‚¬
Gesamtbetrag: 6,74 â‚¬
```

````python
# schreibe deinen code unter diese Zeile


# ProduktwÃ¶rterbuch erstellen
products = {
    "Apfel": 0.75,
    "Banane": 0.99,
    "Brot": 3.49,
    "Milch": 2.50,
    "KÃ¤se": 3.75
}

# Leeren Warenkorb initialisieren
cart = []

# Benutzer nach Artikeln fragen und zum Warenkorb hinzufÃ¼gen
print("VerfÃ¼gbare Produkte:", ", ".join(products.keys()))
for i in range(3):
    item = input(f"Geben Sie Artikel {i+1} ein, um ihn zum Warenkorb hinzuzufÃ¼gen: ")
    if item in products:
        cart.append({"name": item, "price": products[item]})
        print(f"{item} zum Warenkorb hinzugefÃ¼gt.")
    else:
        print(f"Entschuldigung, {item} ist nicht verfÃ¼gbar.")

# Warenkorb und Gesamtbetrag ausgeben
total = 0
print("\nIhr Warenkorb:")
for item in cart:
    print(f"- {item['name']}: {item['price']:.2f} â‚¬")
    total += item['price']

print(f"Gesamtbetrag: {total:.2f} â‚¬")
````

   - Sagen Sie dem Benutzer, ob sein Tipp zu hoch, zu niedrig oder richtig ist
   - Wenn richtig, gratulieren Sie ihm und beenden Sie das Spiel
   - Wenn er alle Versuche aufgebraucht hat, ohne die Zahl zu erraten, geben Sie die Antwort aus
4. Verwenden Sie eine For-Schleife, um die Versuche zu verfolgen
5. Verwenden Sie break, um die Schleife vorzeitig zu beenden, wenn die Zahl richtig erraten wurde

Beispiel:
```
Ich denke an eine Zahl zwischen 1 und 100...
Versuch 1: Dein Tipp? 50

Versuch 2: Dein Tipp? 25

Versuch 3: Dein Tipp? 37

Herzlichen GlÃ¼ckwunsch! Du hast die Zahl in 4 Versuchen erraten!
```

````python
# schreibe deinen code unter diese Zeile


# Die geheime Zahl festlegen
secret_number = 42  # FÃ¼r ein echtes Spiel wÃ¼rde man random.randint(1, 100) verwenden

# Die Rateschleife starten
print("Ich denke an eine Zahl zwischen 1 und 100...")
max_attempts = 10

for attempt in range(1, max_attempts + 1):
    guess = int(input(f"Versuch {attempt}: Dein Tipp? "))
    
    # Jeden Tipp Ã¼berprÃ¼fen
    if guess == secret_number:
        print(f"Herzlichen GlÃ¼ckwunsch! Du hast die Zahl in {attempt} Versuchen erraten!")
        break
    elif guess < secret_number:
        print("Zu niedrig!")
    else:
        print("Zu hoch!")

    # PrÃ¼fen, ob der Spieler alle Versuche aufgebraucht hat
    if attempt == max_attempts:
        print(f"Leider hast du alle {max_attempts} Versuche aufgebraucht. Die Zahl war {secret_number}.")
````

## BonusÃ¼bung 1: Emoji-Ãœbersetzer

Erstellen Sie ein unterhaltsames Emoji-Ãœbersetzer-Programm:

1. Erstellen Sie ein WÃ¶rterbuch, das WÃ¶rtern Emojis zuordnet
   - Beispiel: `{"glÃ¼cklich": "ğŸ˜Š", "traurig": "ğŸ˜¢", "Hund": "ğŸ•", "Katze": "ğŸˆ", ...}`
2. Bitten Sie den Benutzer, einen Satz einzugeben
3. Ersetzen Sie jedes Wort, das ein Emoji-Ã„quivalent hat, durch das Emoji
4. Geben Sie sowohl den ursprÃ¼nglichen Satz als auch den neuen, mit Emojis gefÃ¼llten Satz aus
5. ZÃ¤hlen Sie, wie viele WÃ¶rter ersetzt wurden, und geben Sie diese Zahl aus

Beispiel:
```
Original: Ich bin glÃ¼cklich meine Katze und meinen Hund zusammen spielen zu sehen
Ãœbersetzt: Ich bin ğŸ˜Š meine ğŸˆ und meinen ğŸ• zusammen spielen zu sehen

3 WÃ¶rter wurden durch Emojis ersetzt!
```

Tipp: Verwenden Sie `.replace()`, um Text in Strings durch anderen Text oder Emojis zu ersetzen.

````python
# schreibe deinen code unter diese Zeile

# Emoji-WÃ¶rterbuch erstellen
emoji_dict = {
    "glÃ¼cklich": "ğŸ˜Š",
    "traurig": "ğŸ˜¢",
    "Hund": "ğŸ•",
    "Katze": "ğŸˆ",
    "Liebe": "â¤ï¸",
    "Sonne": "â˜€ï¸",
    "Regen": "ğŸŒ§ï¸",
    "Computer": "ğŸ’»",
    "Buch": "ğŸ“š",
    "Pizza": "ğŸ•"
}

# Satz vom Benutzer holen
original_sentence = input("Gib einen Satz ein: ")
translated_sentence = original_sentence

# WÃ¶rter durch Emojis ersetzen
replaced_count = 0
for word, emoji in emoji_dict.items():
    if word in translated_sentence.lower():
        # Vorkommen vor dem Ersetzen zÃ¤hlen
        word_count = translated_sentence.lower().count(word)
        # Das Wort durch Emoji ersetzen
        translated_sentence = translated_sentence.replace(word, emoji)
        replaced_count += word_count

# Ergebnisse zÃ¤hlen und anzeigen
print(f"\nOriginal: {original_sentence}")
print(f"Ãœbersetzt: {translated_sentence}")
print(f"\n{replaced_count} WÃ¶rter wurden durch Emojis ersetzt!")
````
